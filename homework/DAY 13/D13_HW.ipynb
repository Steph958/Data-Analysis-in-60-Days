### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})

print(df1)
print("\n")
print(df2)
````

````python
# - 依照 fruit 欄位做合併

df_merge = pd.merge(df1, df2, on="fruit")
df_merge
````

````python
# - 依照 index 索引做合併

df_merge = pd.merge(df1, df2, on="index")
df_merge
````
