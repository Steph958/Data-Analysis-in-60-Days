# 1.Matplotlib

```python
#曲線圖

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y,)
```

```python
#subplot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
fig, axes = plt.subplots(2, 2) # 建立 2*2 多維視窗 
data = pd.Series(np.random.rand(5), index=list('12345'))

#ax為選擇畫圖視窗，color為顏色，alpha為透明度設定    
data.plot.bar(ax=axes[0,1], color='b', alpha = 1) 
data.plot.barh(ax=axes[1,1], color='b', alpha=0.5)
data.plot.bar(ax=axes[1,0], color='c', alpha = 0.8) 
data.plot.barh(ax=axes[0,0], color='r', alpha=0.5)
 
plt.show()
```

```python
# 散佈圖

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.scatter(X, Y, color='g',alpha=0.5,s=100,edgecolors='g')
plt.title("Scatter plot ")
```

```python
#長條圖

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(0., 10., 0.7)
y = np.arange(0., 10., 0.7)

plt.bar(x, y)
```

# 2.Pandas

```python
#折線圖

import pandas as pd 
import numpy as np 

df = pd.DataFrame(np.random.randn(10,4),
                  index=pd.date_range('2018/12/18',periods=10),
                  columns=list('ABCD')) 

print(df)
df.plot()
```

```python
import pandas as pd 
import numpy as np 

s = pd.Series(np.random.randn(10).cumsum(),
           index= np.arange( 0, 100, 10)) 

print(s)
s.plot()
```

```python
#長條圖

import pandas as pd 
import numpy as np

fig,axes = plt.subplots(2,1)

data = pd.Series(np.random.rand(16),
              index= list('abcdefghijklmnop'))

print(data)

data.plot(kind='bar', ax=axes[0], color='k', alpha= 0.7) 

data.plot( kind='barh', ax=axes[1], color='k', alpha= 0.7)
```

```python
import pandas as pd 
import numpy as np 

df = pd.DataFrame(np.random.rand(10,4),
                  columns=['a','b','c','d'])

print(df)

df.plot.bar()
```

```python
#累計長條圖

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10,4),
                  columns=['a','b','c','d'])

print(df)

df.plot.bar(stacked=True)
```

```python
#水平長條圖

import pandas as pd
import numpy as np 

df = pd.DataFrame(np.random.rand(10,4),
                  columns=['a','b','c','d']) 

df.plot.barh(stacked=True)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a':np.random.randn(1000)+1,
                   'b':np.random.randn(1000),
                   'c': np.random.randn(1000) - 1},
                  columns=['a', 'b', 'c'])

print(df)
df.plot.hist(bins=40) #使用 plot.hist()方法繪製直方圖，我們可以指定bins的數量值
```

```python
import pandas as pd
import numpy as np

df=pd.DataFrame({'a':np.random.randn(1000)+1,
                 'b':np.random.randn(1000),
                 'c': np.random.randn(1000) - 1},
                columns=['a', 'b', 'c']) 
 
df.hist(bins=20) #為每列繪製不同的直方圖
```

```python
#盒鬚圖

import numpy as np 
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 5),
                  columns=['A', 'B', 'C', 'D', 'E'])

df.plot.box()
```

```python
#區域圖

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd']) 

df.plot.area(color=['purple','pink','orange','yellow'])
```

```python
#散佈圖

import pandas as pd
import numpy as np 

df = pd.DataFrame(np.random.rand(50, 4),
                  columns=['a', 'b', 'c', 'd']) 

print(df.head())

df.plot.scatter(x='a', y='b')
```

```python
#圓餅圖

import pandas as pd
import numpy as np 

df = pd.DataFrame(3 * np.random.rand(4),
                  index=['a', 'b', 'c', 'd'],
                  columns=['x']) 

print(df)

df.plot.pie(subplots=True)
```
